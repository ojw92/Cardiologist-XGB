```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../Data/heart_2022_clean_stepwise0.1.csv")
df.info()
```

```python
# set predict and response variables
set_seed = 123
y = df['HadHeartAttack']
x = df.drop(columns=['HadHeartAttack'])

# train / test data split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=set_seed)
```

```python
# create dataframe for training
train_df = pd.concat([x_train, y_train], axis=1)
train_df = train_df.loc[:, ~train_df.columns.str.contains('^Unnamed', regex=True)]
train_df.head()
```

```python
# create dataframe for testing
test_df = pd.concat([x_test, y_test], axis=1)
test_df = test_df.loc[:, ~test_df.columns.str.contains('^Unnamed', regex=True)]
test_df.head()
```

```python
# generate csv files
train_df.to_csv('../Data/train_df0.1.csv', index = False)
test_df.to_csv('../Data/test_df0.1.csv', index = False)
```
